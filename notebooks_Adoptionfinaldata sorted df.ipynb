```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pet_types= pet2['PetType'].unique()
print(pet_types)
                
```

```python
pet_type_dfs={}

for i in range(len(pet_types)):
    pet_typesnum = pd.DataFrame(columns=(list(pet2.columns)))
    for row in range (len(pet2)):
        if pet2['PetType'].iloc[row] ==  pet_types[i]:
            pet_typesnum.loc[len(pet_typesnum)] = pet2.iloc[row]
    
    pet_type_dfs[pet_types[i]] = pet_typesnum

print(pet_type_dfs)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python

# --- CATEGORICAL EFFECTS (Adoption Rate by Category) ---

categorical_vars = ['PetType', 'Breed', 'Color', 'Size']

for var in categorical_vars:
    plt.figure()
    pet2.groupby(var)['AdoptionLikelihood'].mean().sort_values().plot(kind='bar')
    plt.ylabel('Adoption Rate')
    plt.title(f'Adoption Likelihood by {var}')
    plt.show()


# --- CONTINUOUS EFFECTS (Logistic Regression) ---

X = pet2[['AgeMonths', 'WeightKg', 'AdoptionFee']]
X = sm.add_constant(X)
y = pet2['AdoptionLikelihood']

model = sm.Logit(y, X).fit()
print(model.summary())
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

categorical_vars = ['PetType', 'Breed', 'Color', 'Size']

for var in categorical_vars:
    table = pd.crosstab(pet2[var], pet2['AdoptionLikelihood'])
    chi2, p, dof, ex = chi2_contingency(table)
    print(f"{var}: chi2={chi2:.2f}, p-value={p:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency, binom

categorical_vars = ['PetType', 'Breed', 'Size']

for var in categorical_vars:
    # Group by category and compute mean adoption
    group = pet2.groupby(var)['AdoptionLikelihood']
    means = group.mean()
    counts = group.count()
    
    # 95% confidence interval for proportion using binomial approximation
    ci = 1.96 * np.sqrt((means * (1 - means)) / counts)
    
    # Plot
    plt.figure(figsize=(8,4))
    plt.errorbar(means.index, means, yerr=ci, fmt='o', capsize=5, markersize=8, linestyle='')
    plt.ylabel('Adoption Probability')
    plt.title(f'Adoption Likelihood by {var}')
    plt.xticks(rotation=45)
    plt.ylim(0,1)
    plt.grid(axis='y', linestyle='--', alpha=0.5)
    plt.show()
    
    # Chi-squared test
    table = pd.crosstab(pet2[var], pet2['AdoptionLikelihood'])
    chi2, p, dof, ex = chi2_contingency(table)
    print(f"{var}: chi2={chi2:.2f}, p-value={p:.4f}")
```
